## Select Slow and Fast pokemons

Select those pokemons that are either very slow (with Speed below the bottom 5%) or very fast (Speed above top 95%).
 Store your results in the variable slow_fast_df.

```python
hist = df['Speed'].plot(kind='hist', figsize=(10,5), bins = 100)
hist.axvline(df['Speed'].quantile(0.05), color='red')
hist.axvline(df['Speed'].quantile(0.95), color='green')
```

Pokemons before the red lines are categorized as very slow and after the green lines are very fast

```python
top_5_sp = df['Speed'].quantile(0.95)
Bottom_5_sp = df['Speed'].quantile(0.05)
(top_5_sp, Bottom_5_sp)
```

```python
slow_fast_df = df.loc[(df['Speed'] < Bottom_5_sp) | (df['Speed'] > top_5_sp)]
slow_fast_df.head()
```
